```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv('/content/customer_shopping_behavior.csv')
df.head(12)
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
#change the caseso of column in lower case with under_score
df.columns=df.columns.str.lower().str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.head()
```

```python
# add column in as age group
labels=['young_adult','adult','middle_age','Senier']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
df.head(10)
```

```python
df[['age','age_group']].head()
```

```python
#create a column of parchase frequency days convert from tedt to number
frequency_maping={'Fortnightly':14,'Weekly':7,'Monthly':30,'Quaterly':90,'Bi_weekly':14,'Annually':365}
df['purchase_frequency']=df['frequency_of_purchases'].map(frequency_maping)
df[['purchase_frequency','frequency_of_purchases']].head(10)
```
